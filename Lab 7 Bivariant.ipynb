```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

#load dataset
diamond = pd.read_csv('/content/Diamonds Prices2022.csv')
diamond.head(6)
```

```python
plt.scatter(data=diamond, x='depth',y='carat')
plt.xlabel('Depth')
plt.ylabel('Carat Rating')
```

```python
sns.regplot(data=diamond, x='depth',y='carat')
plt.xlabel('Depth')
plt.ylabel('Carat Rating')
```

```python
sns.regplot(data=diamond, x='x',y='y');
```

```python
#add jitter
sns.regplot(data=diamond, x='price',y='table',x_jitter=500);
```

```python
# Add Transparency
sns.regplot(data=diamond, x='price',y='depth',x_jitter=0.3,
           scatter_kws={'alpha': 1/20});
```

```python
plt.hist2d(data=diamond, x='price',y='depth')
plt.colorbar()
plt.xlabel('Price')
plt.ylabel('Depth')
```

```python
plt.hist2d(data=diamond, x='price',y='carat',cmin=0.5)
plt.colorbar()
plt.xlabel('Price')
plt.ylabel('Carat Rating')
```

```python
base_color = sns.color_palette()[0]
sns.boxplot(data=diamond, x='price', y='depth',color=base_color)
plt.xticks(rotation=15)
```

```python
sns.pointplot(data = diamond, x = 'cut', y = 'price')
plt.ylabel('Acerage cut per price')
plt.xticks(rotation=15)
```
